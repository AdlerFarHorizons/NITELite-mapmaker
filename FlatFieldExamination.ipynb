```python
# read in all the flat frames
DirectoryPath="drive/Shareddrives/NITELite/Data & Analysis/Calibration Tests/3.19.22 Complete Calib Set/23085686_1sec_flatFrames/"

flatFrames=[]
for fName in os.listdir(DirectoryPath):
    if fName.endswith(".raw"):
        # read the image
#        f=open(DirectoryPath+fName,"rb")
        dt = np.dtype(np.int16)
        dt = dt.newbyteorder('<')
        img=np.fromfile(DirectoryPath+fName,dtype=dt)
        img=img.reshape((1200,1920))
        if (np.mean(img)<3500):
          flatFrames.append(img)
          print("Flat Frame: ",fName,np.mean(img))

# average the flat files
#N=len(flatFrames)
#fF=np.stack(flatFrames)
#print(N)
#fF=np.median(fF,axis=0)-dF
#print(fF.shape)
```

```python
# Plot histograms of the red pixel intensities of only one of the flat frames (#3)
# since the raw files are bayer sampled the red pixels are at "[0::2,0::2]" locations, green are at "[0::2,1::2]" and "[1::2,0::2]", and blue at "[1::2,1::2]"

fig = plt.figure(figsize=(15,10))
plt.hist(np.ravel(flatFrames[3][0::2,0::2]),bins=20,histtype="step",color="red")
plt.show()
fig = plt.figure(figsize=(15,10))
plt.hist(np.ravel(flatFrames[3][0::2,0::2]),bins=100,histtype="step",color="red")
plt.show()
fig = plt.figure(figsize=(15,10))
plt.hist(np.ravel(flatFrames[3][0::2,0::2]),bins=500,histtype="step",color="red")
plt.show()
fig = plt.figure(figsize=(15,10))
plt.hist(np.ravel(flatFrames[3][0::2,0::2]),bins=range(2800,3200,1),histtype="step",color="red")
plt.show()
fig = plt.figure(figsize=(15,10))
plt.hist(np.ravel(flatFrames[3][0::2,0::2]),bins=range(2400,2800,1),histtype="step",color="red")
plt.show()
fig = plt.figure(figsize=(15,10))
plt.hist(np.ravel(flatFrames[3][0::2,0::2]),bins=range(3200,3600,1),histtype="step",color="red")
plt.show()
#plt.hist([np.ravel(flatFrames[3][0::2,0::2]),
#          np.ravel(flatFrames[3][1::2,1::2]),
#          np.ravel(flatFrames[3][0::2,1::2]),
#          np.ravel(flatFrames[3][1::2,0::2])], bins=range(2800,3200,1),histtype="step",color=["red","blue","green","green"])
```
